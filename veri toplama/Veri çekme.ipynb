```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
full_urlList = []
fiyatlar_list = []
baslik_list = []
deger_list = []
konumlar_list = []



toplamUrun=0
for sayfaNo in range(1,23):
    url="https://www.emlakjet.com/satilik-daire/mugla/"+ str(sayfaNo)+"/?ilce[]=mugla-marmaris-1517&ilce[]=mugla-koycegiz-1488&ilce[]=mugla-ortaca-1831&ilce[]=mugla-mentese-2089&ilce[]=mugla-ula-1695"  # url for döngüsü içinde güncelleniyor
    r = requests.get(url)
    soup=BeautifulSoup(r.content,"lxml")
    

    evler=soup.find_all("div",attrs={"class":"_3qUI9q"})
    
    
    for ev in evler:
        urunLink = (ev.a.get("href"))  
        full_url = 'https://www.emlakjet.com/' + urunLink # evlerin linkleri bulundu
        full_urlList.append(full_url)  #linkleri kaydediyoruz

        try:
            ev_r = requests.get(full_url)
            toplamUrun +=1
        except Exception:
            print("Ürün detayı alınamadı...")
            
        ev_soap = BeautifulSoup(ev_r.content,"lxml")
         
            
        ozellikler =  ev_soap.find_all("div",attrs={"class":"_35T4WV"})                                                                    
        for ozellik in ozellikler:
            baslik = ozellik.find("div",attrs={"class":"_1bVOdb"}).text
            deger = ozellik.find_all("div",attrs={"class":"_1bVOdb"})[1].text
            print(baslik  +":"+ deger )
            
            baslik_list.append(baslik)
            deger_list.append(deger)
           
       
    fiyatlar = soup.find_all("p",attrs={"class":"_2C5UCT"})
    for fiyat in fiyatlar:                                                                     
        fiyatlar_list.append(fiyat.text)
        
        
    konumlar = soup.find_all("div",attrs={"class":"_2wVG12"})
    for konum in konumlar:                                                                     
        konumlar_list.append(konum.text)
       


    
    
print(str(len(full_urlList))+" Adet link bulundu.")
print(str(len(fiyatlar_list))+" Adet fiyat bulundu.")

```

```python
#Url ve fiyatları bir data frame yazma
df_urls = pd.DataFrame()
df_urls["URLS"] = full_urlList
df_urls["FİYATLAR"] = fiyatlar_list
df_urls["KONUM"] = konumlar_list
```

```python
evler=len(full_urlList)
```

```python
columns = np.array(baslik_list)
columns = np.unique(columns)
```

```python
#Başlıkları kullanarak url ve fiyat ile birlikte yeni bir data frame oluşturma
df = pd.DataFrame(columns=columns)
df["URL"] = full_urlList
df["FİYAT"] = fiyatlar_list
df["KONUM"] = konumlar_list
```

```python
#Oluşturulan data frame gösterme
df.head()
```

```python
#Data frame'i kullanarak bütün verileri çekme ve sütunlara yazdırma
for i in range(0,evler):
    url = df['URL'].loc[i]
    r = requests.get(url)
    soup = BeautifulSoup(r.content,"lxml")
    
    ozellikler = soup.find_all("div",attrs={"class":"_35T4WV"})
    for ozellik in ozellikler:
        baslik = ozellik.find("div",attrs={"class":"_1bVOdb"}).text
        deger = ozellik.find_all("div",attrs={"class":"_1bVOdb"})[1].text
        print(baslik+deger)
        df[baslik].loc[i] = deger
```
